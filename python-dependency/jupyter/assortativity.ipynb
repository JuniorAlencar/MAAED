```python
from src.create_network import *
from src.assortativity import *
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['axes.linewidth'] = 1.4 #set the value globally
```

```python
# Create network
G = create_network()
# Create assortativity file with tuple (degree[i],mean(degree_viz[i]))
df = assortativity_create(G)
x = df["degree"].values
y = df["k_ij"].values
```

```python
num_bins = 15
xbin,ybin = binning_2d(x, y, num_bins)
```

```python
fig, ax = plt.subplots(figsize=(15,10))
ax.plot(xbin,ybin,"o",ms=18,color='red')
ax.set_xlabel(r"$k_i$",fontsize=25)
ax.set_ylabel(r"$\langle k_{ij} \rangle$",fontsize=25)
ax.set_title("Assortativity degree",fontsize=25)
ax.set_xscale("log")
ax.set_yscale("log")
ax.tick_params(which='minor', width=1.4, length=4,direction="in",labelsize=19)
ax.tick_params(which='major', width=1.4, length=8,direction="in",labelsize=19)
plt.savefig("../results/assortativity_degree.pdf",dpi=300)
plt.show()
```
