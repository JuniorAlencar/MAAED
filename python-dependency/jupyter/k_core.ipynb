```python
# create_network define in src/create_network.py
G = create_network()
```

```python
k = 3  # Defina o valor de k desejado

# Calcule o k-core
k_core = nx.k_core(G, k)
print(k_core)
```

```python
k_core_sizes = []
k_values = range(1, 11)  # Vamos calcular para k de 1 a 10

for k in k_values:
    k_core = nx.k_core(G, k)
    k_core_sizes.append(len(k_core))

# Crie o gráfico
plt.ylabel(r"$\langle k \rangle$",fontsize=18)
plt.xlabel(r"$k_{core}$",fontsize=18)

plt.plot(k_values, k_core_sizes,"o")
```

```python
plt.figure(figsize=(16,8))
ms= 12
plt.plot(k_values, k_core_sizes,'o',ms=ms,mec=None,markerfacecolor='w')
plt.xscale("log")
plt.yscale("log")
plt.ylabel("Number of nodes",fontsize=20)
plt.xlabel(r"$k_{core}$",fontsize=20)
plt.xticks(fontsize=12)  
plt.yticks(fontsize=12)  
plt.show()
```

```python
k_medio_values = []

for k in k_values:
    k_core2 = nx.k_core(G, k)
    
    # Calcule o valor médio de k para o k-core
    k_core_degrees = [d for n, d in k_core2.degree()]
    k_medio = sum(k_core_degrees) / len(k_core_degrees)
    k_medio_values.append(k_medio)

fig = plt.figure(figsize=(18, 10),edgecolor='black', linewidth=1.4)  
plt.plot(k_values, k_medio_values, marker='o', linestyle='-',ms=18,mcf='none')
plt.xlabel(r'$k_{core}$',fontsize=25)
plt.ylabel(r'$\langle k \rangle$',fontsize=25)
plt.tick_params(axis='both', which='both', labelsize=18)
plt.tick_params(axis='both', which='both', direction="in",length=7,width=1.4)
#plt.title('Gráfico k_médio vs. k_core')
#plt.grid(True)
plt.show()
```
